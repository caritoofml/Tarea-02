```python
# Problema 4: Calculador de perfil musical
# Este programa evalúa el perfil de una persona según su edad y preferencias,
# generando una clasificación simple que interpreta su relación con la música
# y el baile en un contexto social.

# Solicitar datos al usuario
edad = int(input("Ingresa tu edad: "))

# Se pide responder con si o no
# Se utiliza .lower() para aceptar respuestas como "Si", "SI" o "si" y también
# "No", "NO", "no", evitando errores por mayúsculas/minúsculas
le_gusta_musica = input("¿Te gusta la música? (si/no): ").lower() == "si"
le_gusta_bailar = input("¿Te gusta bailar? (si/no): ").lower() == "si"

# Evaluar perfil del usuario
if edad >= 18 and le_gusta_musica and le_gusta_bailar:
    # Perfil social activo y participativo
    print("Eres una estrella de la fiesta")

elif le_gusta_musica and not le_gusta_bailar:
    # Disfruta la música, pero sin interés en el baile
    print("Te gusta la música, pero no tanto bailar")

else:
    # Caso general: preferencias diversas
    print("Tienes tu propio estilo")
```
