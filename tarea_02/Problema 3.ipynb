```python
# Problema 3: Detector de puntaje gamer
# Este programa recoge dos puntajes ingresados por el usuario y los analiza en conjunto.
# A partir del resultado total, se clasifica el nivel de desempeño en categorías simples,
# permitiendo una lectura rápida del rendimiento.

# Solicitar puntajes al usuario (convertidos a número entero)
puntaje1 = int(input("Ingresa primer puntaje: "))
puntaje2 = int(input("Ingresa segundo puntaje: "))

# Sumar ambos puntajes para obtener el resultado total
total = puntaje1 + puntaje2

# Evaluar el nivel según el puntaje obtenido
if total < 50:
    # Rendimiento inicial
    print("Resultado: Nivel principiante")

elif total < 80:
    # Nivel medio de desempeño
    print("Resultado: Nivel intermedio")

else:
    # Alto rendimiento
    print("Resultado: Nivel pro")
```
