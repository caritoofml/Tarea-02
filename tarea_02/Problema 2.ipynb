```python
# Problema 2: Números especiales del robot
# Este programa simula un sistema de identificación de patrones numéricos,
# donde se analizan valores del 1 al 100 para detectar aquellos que cumplen
# ciertas condiciones matemáticas. En este caso, se identifican números
# divisibles por 5 y se destacan especialmente aquellos que también son
# divisibles por 10, asignándoles una señal diferenciada.

for numero in range(1, 101):

    # Los números divisibles por 10 representan un caso particular,
    # ya que cumplen con ambas condiciones (5 y 10), por lo que reciben
    # una señal doble.
    if numero % 10 == 0:
        print(numero, "Bip Bip")

    # En segundo lugar, se identifican los números que cumplen
    # únicamente con la condición de ser divisibles por 5.
    elif numero % 5 == 0:
        print(numero, "Bip")
```
